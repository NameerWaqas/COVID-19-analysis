# Visualizing Data

### Number of Patients vs Date of report 

```python
df.plot(kind='bar',x='Date',y='Pakistan',figsize=(20,10),grid=True,color='r')
plt.title('Daily number of patients reported by date in Pakistan')
plt.xlabel('DATE')
plt.ylabel('Number of patients')
plt.grid(color='g')
plt.savefig(r'D:\CoronaAnalysis\24MAR20\DailyReportedPatients.png')
```

### Commulative sum of Patients vs Date

```python
cumSum = np.cumsum(df['Pakistan'])
```

```python
df['cumSumPak'] = cumSum
df
```

```python
df.plot(kind='line',x='Date',y='cumSumPak',figsize=(20,10),grid=True,color='r',label='Pakistan')
plt.scatter(x=df['Date'],y=df['cumSumPak'],color='r')
plt.xticks(rotation='vertical')
plt.grid(which='major',color='g')
plt.title('Cummulative patients by date in Pakistan')
plt.xlabel('DATE')
plt.ylabel('Number of patients')
plt.savefig(r'D:\CoronaAnalysis\24MAR20\CmPatientsPak.png')
```
