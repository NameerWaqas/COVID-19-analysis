```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Population of Countires

```python
Pakistan = 220892340
China = 1439323776
Iran = 83992949
Italy = 60461826
US = 331002651
```

# Importing DATASET

```python
df = pd.read_excel(r'C:\Users\NameerWaqas\Downloads\WHO_Data_Corona_VirusDataForEda.xlsx')
df.drop(59,inplace=True)
tempDF = pd.DataFrame([["22MAR2020",83,6600,2000,4800,np.nan,55,np.nan,np.nan],["22MAR2020",104,5600,1400,16400,np.nan,109,np.nan]],columns=df.columns)
df = pd.concat([df,tempDF],ignore_index=True)
df
```

```python
df.head(100)
```

```python
df['China'].replace(0,126,inplace=True)
df[['Pakistan','Italy','US','Iran']]=df[['Pakistan','Italy','US','Iran']].replace(np.nan,0.1)
df
```

```python
df.describe()
```

```python
df['PakistanR'] = df['Pakistan']/Pakistan
df['ChinaR'] = df['China']/China
df['ItalyR'] = df['Italy']/Italy
df['IranR'] = df['Iran']/Iran
df['USR'] = df['US']/US
```

```python
df.describe()
```

# Formatting Date

```python
formatedDate= list()
for i in range(23,32,1):
    formatedDate.append(str(i)+"JAN"+"2020")
for i in range(1,30,1):
    formatedDate.append(str(i)+"FEB"+"2020")
for i in range(1,24,1):
    formatedDate.append(str(i)+"MAR"+"2020")
```

```python
formatedDate
```

```python
df.Date.replace(df.Date.values,formatedDate,inplace=True)
```

```python
df.Date[21]="13FEB2020"
df.Date[21]
```

```python
x=df.Date
fig = plt.figure()
ax = fig.add_axes([0,0,2.5,2])
ax.plot(x,df['China'],label='China')
ax.plot(x,df['Pakistan'],label='Pakistan')
ax.plot(x,df['US'],label='US')
ax.plot(x,df['Iran'],label='Iran')
ax.plot(x,df['Italy'],label='Italy')
ax.set_xlabel("Day")
ax.set_ylabel("Number of inefcted people")
ax.set_title("Number of daily effected people in Pakistan corresponding to other countries")
plt.xticks(rotation='vertical')
# plt.yticks(ticks=range(0,16000,500))
plt.grid(which='major')
plt.legend()
plt.show()
# plt.savefig(r'C:\Users\NameerWaqas\Desktop\dailyPakVsOther.png')
# ax = df.plot(kind='line',x = 'Date',y=['China','Pakistan','Italy','US','Iran'],figsize=(20,10),grid=True)
```
