# Visualizing Data

### Number of Patients vs Date of report 

```python
df.plot(kind='bar',x='Date',y='Pakistan',figsize=(20,10),grid=True,color='r',fontsize=12)
plt.title('Daily number of patients reported by date in Pakistan',fontsize=20)
plt.xlabel('DATE',fontsize=20)
plt.ylabel('Number of patients',fontsize=20)
plt.grid(color='g')
plt.savefig(r'D:\CoronaAnalysis\30MAR20\DailyReportedPatients.png')
```

### Commulative sum of Patients vs Date

```python
cumSum = np.cumsum(df['Pakistan'])
```

```python
df['cumSumPak'] = cumSum
df
```

```python
df.plot(kind='line',x='Date',y='cumSumPak',figsize=(20,15),grid=True,color='r',label='Pakistan',fontsize=12)
# fig,ax=plt.subplots(figsize=(20,10))
plt.scatter(x=df['Date'],y=df['cumSumPak'],color='r')
plt.xticks(rotation='vertical')
plt.yticks(ticks=range(0,1800,100))
plt.grid(which='major',color='g')
plt.title('Cummulative patients by date in Pakistan',fontsize=20)
plt.xlabel('DATE',fontsize=20)
plt.ylabel('Number of patients',fontsize=20)
for i,j in enumerate(df['Pakistan']):
    if i>=53:
        plt.annotate(" "+df['Date'][i]+": "+str(int(j)),(df['Date'][i],df['cumSumPak'][i]))
plt.savefig(r'D:\CoronaAnalysis\30MAR20\CmPatientsPak.png')
```
